## Предобработка данных

```python
import numpy as np
# Для начала tbd обозначу как Nan и поменяю тип
games.loc[games.user_score == 'tbd','user_score'] = float('nan') 
games['user_score'] = games['user_score'].astype(float)

# удаляю пустые жанры
games.dropna(inplace = True, subset = ['name']) 

# заполняю пустые года по выпуску на разных платформах
games['year_of_release'] = np.round(games['year_of_release'].where(~games['year_of_release'].isnull(), games.groupby('name')["year_of_release"].transform('mean')))
# что не получилось - обнуляю
games.loc[games.year_of_release.isnull(),'year_of_release'] = 0
games['year_of_release'] = games['year_of_release'].astype(int)

# заполняю пустые оценки и рейтинг
games['user_score_without_na'] = np.round(games['user_score'].where(~games['user_score'].isnull(), games.groupby('name')["user_score"].transform('mean')),2)
games.loc[games.user_score_without_na.isnull(),'user_score_without_na'] = 0
games.loc[games.user_score.isnull(),'user_score'] = 0

games['critic_score_without_na'] = np.round(games['critic_score'].where(~games['critic_score'].isnull(), games.groupby('name')["critic_score"].transform('mean')),2)
games.loc[games.critic_score_without_na.isnull(),'critic_score_without_na'] = 0
games.loc[games.critic_score.isnull(),'critic_score'] = 0

# Заполняю самым частым рейтингом для жанра, если для конкретной игры не нашлось
func = lambda s: s.values[0]
games['rating_without_na'] = games['rating'].where(~games['rating'].isnull(), games.groupby('name')["rating"].transform(func))

func = lambda s: s.value_counts().idxmax() if len(s.value_counts())>0 else float('Nan')
games['rating_without_na'] = games['rating_without_na'].where(~games['rating_without_na'].isnull(), games.groupby(['year_of_release','genre'])["rating"].transform(func))
games.loc[games.rating_without_na.isnull(),'rating_without_na'] = 'u'

games.loc[games.rating.isnull(),'rating'] = 'u'


# Обработка дублей
dubl = games[games.duplicated(subset = ['name', 'platform', 'year_of_release'])]
s = games.merge(dubl, on=['name', 'platform', 'year_of_release'], suffixes = ('','_'))
if s.shape[0] == 0: 
    pass
else:
    # Так как это не полные дубли, а одно и тоже, но почему то разнесённое по разным строкам
    # то нужно сложить цифровые значения в разрезе дублей.
    q = s[list(games.columns)].groupby(['name', 'platform', 'year_of_release']).agg({'genre':'max',
                                                                                       'na_sales':'sum',
                                                                                       'eu_sales':'sum',
                                                                                       'jp_sales':'sum',
                                                                                       'other_sales':'sum',
                                                                                       'critic_score':'max',
                                                                                       'user_score':'max',
                                                                                       'rating':'max'
                                                                                       }).reset_index()  
    games.drop_duplicates(subset=['name', 'platform', 'year_of_release'], keep=False, inplace=True)
    games= pd.concat([games,q],ignore_index=True, sort=False)
    
games.info()
```

```python
# Посчитайте суммарные продажи во всех регионах и запишите их в отдельный столбец.
games['all_sales'] = games['na_sales'] + games['eu_sales'] + games['jp_sales'] + games['other_sales']
```

```python
def create_diag(platform, mas, dob):
    # Убираю выбросы по продажам, по оценке критиков, по оцнеке пользователей 
    df = games_after_2013[(games_after_2013.platform == platform)].copy() 
    for i in mas:
        q =  df[df[i]>0][i].describe()
        q1 = q['25%']
        q3 = q['75%']
        mq = q3 - q1
        filter_up =  (df[i] > (q3 + mq*1.5))
        filter_down = (0 < df[i]) & (df[i] < (q1 - mq*1.5))
        #обнуляю
        df.loc[filter_up, i] = 0
        df.loc[filter_down, i] = 0
    # Удаляю строки, где хотя бы одно из этих значений 0
    df.drop(df[(df[mas[0]] == 0)|(df[mas[1]] == 0)|(df[mas[2]] == 0)].index, inplace=True)

    sns.scatterplot(x="all_sales", y=mas[1],  linewidth=0,  data=df)
    plt.title("Диаграмма разброса "+ platform + dob, fontsize=18, fontstyle="italic", fontweight="bold", linespacing=10)
    plt.xlabel('Продажи на ' + platform)
    plt.ylabel('Оценка пользователей')
    plt.show()

    sns.scatterplot(x="all_sales", y=mas[2],  linewidth=0,  data=df)
    plt.xlabel('Продажи на ' + platform)
    plt.ylabel('Оценка критиков')
    plt.show()

    display(df[mas].corr())

create_diag('PS4',['all_sales','user_score','critic_score'], ' (с незаполненым средним)')
create_diag('PS4',['all_sales','user_score_without_na','critic_score_without_na'], ' (с заполненым средним)')
```

```python
create_diag('3DS',['all_sales','user_score','critic_score'], ' (с незаполненым средним)')
create_diag('3DS',['all_sales','user_score_without_na','critic_score_without_na'], ' (с заполненым средним)')
```

```python
create_diag('XOne',['all_sales','user_score','critic_score'], ' (с незаполненым средним)')
create_diag('XOne',['all_sales','user_score_without_na','critic_score_without_na'], ' (с заполненым средним)')
```

Можно сказать что отзывы критикиков и отзывы игроков коррелируют от слабо, до умеренно. Оценка пользователей на продажи либо слабо влияет, либо влияние нелинейно. Оценка критиков влияет от вовсе нет, до умеренно. Всё слишком сильно зависит от платформы, так как у разных платформ разные собщества, со своими нюансами, историей, культурой. ## Проверка гипотез

```python

from scipy import stats as st
alpha = 0.01

# Нулевая гипотеза: Средние пользовательские рейтинги платформ Xbox One и PC не отличаются
# Альтернативная гипотеза: Средние пользовательские рейтинги платформ Xbox One и PC отличаются
# t-тест можно делать если:
# Сравниваемые значения не составляют пару коррелирующих значений
# Дисперсии признака в выборках примерно равны
# нормальное распределение на больших выборках не является обязательным условиям
# корреляции тут явно нет, так как априори независимые выборки
# Дисперсию проверю по критерию Фишера

d1 = games_after_2013.query('(platform == "XOne") & (user_score>0)').user_score        
d2 = games_after_2013.query('(platform == "PC")  & (user_score>0)').user_score        
print('Критерий Фишера:',np.var(d2)/np.var(d1))
print('Размерность: ' + str(len(d1)-1)+'x'+str(len(d2)-1))
```

```python
# По таблице при уровне значисмости 0,01 данный показатель означает, что дисперсии отличаются.
# В таком случае нельзя делать просто t тест. Можно сделать t-критерий Уэлча 
results =  st.ttest_ind(d1, d2, equal_var=False)

print('p-value: '+ str(results.pvalue))
if results.pvalue < alpha:
    print('Средние пользовательские рейтинги отличаются')
else:
    print('Средние пользовательские рейтинги не отличаются')
```

```python
d1 = games_after_2013.query('(platform == "XOne") & (user_score_without_na>0)').user_score_without_na        
d2 = games_after_2013.query('(platform == "PC")  & (user_score_without_na>0)').user_score_without_na        
print('Критерий Фишера:',np.var(d2)/np.var(d1))
print('Размерность: ' + str(len(d1)-1)+'x'+str(len(d2)-1))
results =  st.ttest_ind(d1, d2, equal_var=False)

print('p-value: '+ str(results.pvalue))
if results.pvalue < alpha:
    print('Средние пользовательские рейтинги отличаются')
else:
    print('Средние пользовательские рейтинги не отличаются')
```
